```python
import openpyxl
import pandas as pd
import string
import math
```

```python
def resize_column(Dataframe_list, sheet_name_list, export_folder):
    
    def Letters(nb_columns): # Fonction pour convertir un nombre en nom de colonne lettre
            res1 = round(nb_columns/26)
            if nb_columns<27:
                result=str(string.ascii_uppercase[nb_columns-1]) + str(1)
            else:
                first_letter = str(string.ascii_uppercase[res1-1])
                second_letter = str(string.ascii_uppercase[nb_columns -1- res1*26])
                result=first_letter + second_letter + str(1)
            return result
         # Create a Pandas Excel writer using XlsxWriter as the engine.
    
    writer = pd.ExcelWriter(export_folder, engine='xlsxwriter', datetime_format='dd/mm/yyyy')       
    for j, df in enumerate(Dataframe_list):
        maxColumnLenghts = [] # Max columns lenghts values
        col_lenght = [] # Max columns value
        new_col_lenght = []

        for col in range(len(df.columns)):
            maxColumnLenghts.append(max(df.iloc[:,col].astype(str).apply(len)))

        for col in df.columns:
            col_lenght.append(len(col))

        for i, col_len in enumerate(col_lenght):
            if col_len >= maxColumnLenghts[i]:
                if col_len>=35:
                    new_col_lenght.append(35)
                else:
                    new_col_lenght.append(col_len+2)
            else:
                if maxColumnLenghts[i]>=35:
                    new_col_lenght.append(35)
                else:
                    new_col_lenght.append(maxColumnLenghts[i]+3)

        # Create a Pandas Excel writer using XlsxWriter as the engine.
         #writer = pd.ExcelWriter(export_folder, engine='xlsxwriter', datetime_format='dd/mm/yyyy')

        # Convert the dataframe to an XlsxWriter Excel object.
 
        df.to_excel(writer, sheet_name=sheet_name_list[j], index=False)

        # Get the xlsxwriter workbook and worksheet objects.
        workbook  = writer.book
        worksheet = writer.sheets[sheet_name_list[j]]
        
        for i, col_len in enumerate(new_col_lenght):
            worksheet.set_column(i, i, col_len)
           

        headerformat= workbook.add_format()       # Set properties 
        headerformat.set_font_color('white')
        headerformat.set_font_size(30)
        headerformat.set_bold()
        headerformat.set_bg_color('navy')
        headerformat.set_border(style=2)
        headerformat.set_border_color('white')
        headerformat.set_text_wrap()
        Last_letter = Letters(len(new_col_lenght))

        range_col = "A1:" + Last_letter

        worksheet.conditional_format(range_col, {'type': 'no_errors',
                                                  'format': headerformat})

        #worksheet.set_row(0, 25)
        worksheet.freeze_panes(1, 0)

    writer.save()
        
  
```
